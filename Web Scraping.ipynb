```python
import requests
```

```python
url_to_scrape = "https://www.hamropatro.com/rashifal"
response = requests.get(url_to_scrape)
```

```python
print(response.status_code)
```

```python
#response.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content,"html5lib")
```

```python
print(soup.prettify())
```

```python
rashifals = soup.find_all("div",attrs={'class':'item'})
```

```python
print(rashifals)
```

```python
print(len(rashifals))
```

```python
for rash in rashifals:
    print(rash.h3.text)
    print(rash.p.text)
    print(rash.img["src"])
    title = rash.h3.text
    description = rash.p.text
    
```

```python
rashifal_dict = {}
for rash in rashifals:
    title = rash.h3.text
    description = rash.p.text
    rashifal_dict[title]=description
    
print(rashifal_dict)
```
